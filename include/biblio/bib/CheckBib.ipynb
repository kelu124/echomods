```python
import bibtexparser
import os.path

INVIS = "[](@autogenerated - invisible comment)"
with open('all.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

#print(bib_database.entries)
```

```python
print(bib_database.comments)
print(bib_database.preambles)
print(bib_database.strings)
```

# DOIs but no PDF found

```python
AllRefs = ""
AllGoodTitles = [""]
for k in bib_database.entries:
    #print k.keys()
    m = "* "
    if os.path.isfile("ArticlesPDF/"+k["ID"]+".pdf"):
        m += "[x] "
    else:
        m += "[ ] "
    m += "["+k["ID"]+"] __"+k["title"]+"__"
    AllGoodTitles.append(k["title"].lower().replace("{","").replace("}",""))
    if "author" in k.keys():
        m += " by "+k["author"]
   
    if "doi" in k.keys():
        m += ". (_DOI: ["+k["doi"]+"](https://doi.org/"+k["doi"]+")_)."
        AllGoodTitles.append(k["doi"])
        if not os.path.isfile("ArticlesPDF/"+k["ID"]+".pdf"):
            print(m.replace("{","").replace("}",""))
    AllRefs += m+"\n"
```

```python
text_file = open("AllRefs.md", "w")
text_file.write(str(AllRefs).replace("{","").replace("}","")+"\n"+str(INVIS))
text_file.close()
```

```python
AllRefs = ""
for k in bib_database.entries:
    #print k.keys()
    m = "* "
    if os.path.isfile("ArticlesPDF/"+k["ID"]+".pdf"):
        m += "[x] "
    else:
        m += "[ ] "
    m += "["+k["ID"]+"] __"+k["title"]+"__"
    if "author" in k.keys():
        m += " by "+k["author"]
   
    if "doi" not in k.keys():
        if not os.path.isfile("ArticlesPDF/"+k["ID"]+".pdf"):
            print(m.replace("{","").replace("}",""))
    AllRefs += m+"\n"
```

```python
devices = "sharmistha adl5511 AFE5809 AD8331 ad8332 NMT0572SC LM96511 AD9276 AD9273 LM96550 MAX668 MAX1856 MIC3172 HV9150 MAX15031 ADA4807 MAX14808 AD9276 AD9273 AFE5808 max2077 MAx2082  AFE5851 MAX4107 AD9272 ADC10D020 AD9276 VCA8500 AFE5809 AFE5808"
devices += " LM96530 XTR-2020 MAX1438 MAX2079 MAX2084 MAX14813 MAX4936 STHV748 AD9671 AD8334 max14866 MD2134 MD0101 MD0100 HV7360 HV7361 HV7321 AFE5809 LM96551 LM96530 LM96570 AD9276"
devices += " hv2605 HV7351 verasonics lecoeur" #Cortex-A9 Altera xilinx
devices += " sthv800 hv20220 biosono ADA4807 Imasonic" 
devices += " MAX14808 TX810 AFE5851 LM96570 HV7350 LM96530 LM96570 AD8031 AD9214 ADS5517 SMA231 THS4509 LT3958 ADC10065 iCE40HX4K MD2131 AFE5805 AD9276 MD1210 TC6320"
devices = list(set(devices.replace("  "," ").lower().split(" ")))
```

```python
import os,glob
result = {}

graph = {}


for dev in devices:
    result[dev] = []
folder_path = './ArticlesPDF/txt/'

for filename in glob.glob(os.path.join(folder_path, '*.txt')):
    with open(filename, 'r') as f:
        text = f.read().lower().replace("max-",'max').replace("xtr-",'xtr').replace("ad-",'ad').replace("afe-",'afe').replace("hv-",'hv')
        text = ' '.join(text.split()).strip()
        text = text.replace("doi: ","doi:")
        fn = filename.split("/")[-1].split(".")[0]
        graph[fn] = []
        for m in AllGoodTitles:
            if m in text:
                graph[fn].append(m)
        lst = text.split(" ")
        for o in lst:
            if o.startswith("doi:") or o.startswith("10.1"):
                o = o.replace("http://doi.org/","").replace("http://dx.doi.org/","").replace("https://doi.org/","").replace("https://doi.","")
                o = o.replace("https://","").replace("doi:","").replace("http://dx.","")
                #print("* "+o.rstrip())
                graph[fn].append(o.rstrip())
        #print(filename)
        for dev in devices:
            if len(dev) and dev in text:
                result[dev].append(filename.split("/")[-1].split(".")[0])
print(result)
```
